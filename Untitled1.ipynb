```python
import numpy as np
def tanh(x):
    return np.tanh(x)
i1 = 0.05
i2 = 0.10
b1 = 0.5
b2 = 0.7
np.random.seed()
w_ih = np.random.uniform(-0.5, 0.5, (2, 2))
w_ho = np.random.uniform(-0.5, 0.5, (2, 2))
h1_input = i1 * w_ih[0][0] + i2 * w_ih[1][0] + b1
h2_input = i1 * w_ih[0][1] + i2 * w_ih[1][1] + b1
h1_output = tanh(h1_input)
h2_output = tanh(h2_input)
o1_input = h1_output * w_ho[0][0] + h2_output * w_ho[1][0] + b2
o2_input = h1_output * w_ho[0][1] + h2_output * w_ho[1][1] + b2
o1_output = tanh(o1_input)
o2_output = tanh(o2_input)
print("Hidden Layer Outputs:")
print("h1 =", h1_output)
print("h2 =", h2_output)
print("\nFinal Outputs:")
print("o1 =", o1_output)
print("o2 =", o2_output)
```
